```python
import osmnx as ox
import geopandas as gpd
import hvplot.pandas
```

```python
place = "Oglala Lakota College, South Dakota, USA"
olc_boundary = ox.geocode_to_gdf(place)
olc_boundary
```

```python
olc_boundary.hvplot(
    geo=True,
    tiles="EsriImagery",
    alpha=0.3,
    line_color="black",
    line_width=2,
    title="Oglala Lakota College Boundary",
    width=800,
    height=600
    
)
```
