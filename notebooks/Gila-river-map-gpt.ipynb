```python
import osmnx as ox
import geopandas as gpd
import hvplot.pandas
```

```python
place = "Gila River Indian Community, Arizona, USA"
gilaboundary = ox.geocode_to_gdf(place)
gilaboundary
```

```python
gilaboundary.hvplot(
    geo=True,
    tiles="EsriImagery",
    alpha=0.3,
    line_color="black",
    line_width=2,
    title="Gila River Indian Community Boundary",
    width=800,
    height=600
    
)
```
